```python
import pandas as pd
import sqlite3 as sql
```

```python
conn = sql.connect("stores.db")
```

```python
rq = pd.read_sql
```

```python
q = """

Select * from sqlite_master

 where type = 'table'

"""

tables = rq(q,conn)

tables
```

```python
table_name = tables["name"]
table_name.tolist()
```

```python
table_cols = tables["sql"].str.extractall(r"\n\s+(\w+)")

bol = table_cols[0].str.contains(r"PRIMARY|FOREIGN")

table_cols = table_cols[~bol].reset_index()

n_cols = table_cols.groupby("level_0").size()
n_cols.tolist()
```

```python
q = ""

for i, c_name, n_col in zip(range(0,8),table_name,n_cols):
    if i ==0:
        q = f"""
        Select '{c_name}' table_name, {n_col} number_of_attributes, count(*)  number_of_rows
            from {c_name} """
    else:
        q += f"""
        Union
            Select '{c_name}' table_name, {n_col} number_of_attributes, count(*)  number_of_rows
            from {c_name} 
        """
        
rq(q,conn)
```

### Which Products Should We Order More of or Less of?



```python
# Low stock

sq = """select quantityInStock from products p where p.productCode = od.productCode"""

q = f"""
Select productCode, round(sum(quantityOrdered)*1.0/({sq}),2) low_stock from

orderdetails od

group by 1

order by 2 Desc

limit 10

"""


rq(q,conn)
```

```python
# product performance

# Low stock


q = f"""
Select productCode, sum(quantityOrdered*priceEach) as prod_perf

 From orderdetails 

group by 1

order by 2 Desc

limit 10

"""


rq(q,conn)
```

```python
sq = """select quantityInStock from products p where p.productCode = od.productCode"""

q = f"""

With 

low_stock_table as 

(Select productCode, round(sum(quantityOrdered)*1.0/({sq}),2) low_stock from

orderdetails od

group by 1

order by 2 Desc

limit 10)

Select productCode, sum(quantityOrdered*priceEach) as prod_perf

 From orderdetails 
 
 where productCode in (select productCode from low_stock_table)

group by 1

order by 2 Desc

limit 10



"""


rq(q,conn)
```
