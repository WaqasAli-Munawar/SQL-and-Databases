```python
import pandas as pd
import sqlite3 as sql
```

```python
conn = sql.connect("ksprojects.db")
```

```python
def read_query(q):
    return pd.read_sql(q,conn)
```

```python
tables = "select * from sqlite_master where type = 'table'"

read_query(tables)
```

```python
# Retrieving Column Data Types

table_info = "Pragma table_info(ksprojects)"

read_query(table_info)
```

```python
# Initial Selection of Rows and Columns

read_query("""SELECT main_category, goal, backers, pledged
  FROM ksprojects
 LIMIT 10;""")
```

```python
# Filtering by Category

read_query("""SELECT main_category, backers, pledged, goal
 FROM ksprojects
WHERE state IN ('failed', 'canceled', 'suspended')
LIMIT 10;""")
```

```python
# Filtering by Quantity

read_query("""SELECT main_category, backers, pledged, goal
  FROM ksprojects
 WHERE state IN ('failed', 'canceled', 'suspended')
   AND backers >= 100 AND pledged >= 20000
 LIMIT 10;""")
```

```python
# Applying Conditional Logic

read_query("""SELECT main_category, backers, pledged, goal, 
       pledged/goal AS pct_pledged,
       CASE
       WHEN pledged/goal  >= 1 THEN "Fully funded"
       WHEN pledged/goal BETWEEN .75 AND 1 THEN "Nearly funded"
       ELSE "Not nearly funded"
       END AS funding_status 
  FROM ksprojects
 WHERE state IN ('failed')
       AND backers >= 100 AND pledged >= 20000
 ORDER BY main_category, pct_pledged DESC
 LIMIT 10;""")
```
