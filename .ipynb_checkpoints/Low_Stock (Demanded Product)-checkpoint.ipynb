```python
import pandas as pd

import sqlite3 as sql
```

```python
db_file = "stores.db"

conn = sql.connect(db_file)
```

```python
# pd.read_sql(query,conn) -> rq(query,conn)

rq = pd.read_sql
```

```python
q = """
select * from sqlite_master

"""

rq(q, conn)
```

```python
q = """
select * from sqlite_master

where type = 'table'

"""

rq(q, conn)
```

# Product Analysis using SQL & Pandas 


## 1. Low Stock (Demanded Product) -> quantity ordered/ quantity stock

```python
q = """

select * from orderdetails

limit 10

"""

rq(q,conn)
```

```python
q = """

SELECT productCode, quantityOrdered

FROM orderdetails


"""

rq(q,conn)
```

```python
q = """

SELECT productCode, sum(quantityOrdered)

FROM orderdetails

group by productCode


"""

rq(q,conn)
```

```python
q = """

SELECT productCode, sum(quantityOrdered)/(select quantityInStock from products as p where p.productCode = od.productCode)

FROM orderdetails as od

group by productCode


"""

rq(q,conn)
```

```python
q = """

SELECT productCode, sum(quantityOrdered)*1.0/(select quantityInStock from products as p 
                                                        where p.productCode = od.productCode) as low_stock

FROM orderdetails as od

group by productCode


"""

rq(q,conn)
```

```python
q = """

SELECT productCode, sum(quantityOrdered)*1.0/(select quantityInStock from products as p 
                                                        where p.productCode = od.productCode) as low_stock

FROM orderdetails as od

GROUP BY productCode

ORDER BY 2 DESC

"""

rq(q,conn)
```

```python
q = """

SELECT productCode, sum(quantityOrdered)*1.0/(select quantityInStock from products as p 
                                                        where p.productCode = od.productCode) as low_stock

FROM orderdetails as od

GROUP BY productCode

ORDER BY 2 DESC

LIMIT 10
"""

rq(q,conn)
```
